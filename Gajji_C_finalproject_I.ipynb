#### Who are the customers that are likely to convert to priemer?

```python
df_corr = df.corr()
df_corr['ispremier'].sort_values(ascending=False)
```

The `ispremier` variable, which indicates Premier membership, shows the highest positive correlations with spending-related features: `conc_bal` (0.3446), `concloydisc` (0.3355), `concqtysum` (0.3353), `ticket_qty` (0.2604), `life_spend_amt` (0.2203), `life_award_amt` (0.2163), and `ticket_bal` (0.2008), suggesting that higher spending and engagement in ticket and concession purchases are linked to Premier status. Moderate correlations are observed with estimated income (`est_inc_rng` and `unk_inc`) and home value ranges (`est_home_val_rng` and `unk_home`), indicating a slight tendency for members with higher economic standing to be Premier members. Opt-in behaviors like `theatr_mob_offer` and `reward_mobile` show low positive correlations, while younger age groups (`age13to18` and `age18to21`) and having children (`child_pres`) are negatively correlated with Premier membership. `gendercode` has a notable negative correlation (-0.2576), implying that female members are less likely to upgrade to Premier status. Overall, spending behavior and economic indicators are the strongest positive predictors of Premier membership.

```python
# Plot heatmap for correlations with ISPREMIER
plt.figure(figsize=(4, 8))
sns.heatmap(df_corr[['ispremier']].sort_values(by='ispremier', ascending=False), annot=True, cmap='coolwarm')
plt.title("Correlation of Features with Premier Status")
plt.show();
```

```python
# Compare age distribution by Premier status
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', hue='ispremier', kde=True, bins=30)
plt.title("Age Distribution by Premier Status")
plt.xlabel("Age")
plt.ylabel("Density")
plt.show();
```

The correlation between age and `ispremier` suggests a slight positive relationship, with older age groups being marginally more likely to hold Premier membership. Specifically, `age` and `age21plus` have low positive correlations (0.0639 and 0.0601, respectively) with `ispremier`, indicating that Premier members tend to be slightly older. Conversely, the younger age categories `age13to18` and `age18to21` show weak negative correlations (-0.0548 and -0.0551, respectively), implying that younger members are slightly less likely to upgrade to Premier. Overall, while age does not strongly predict Premier status, there is a minor tendency for older members to be more inclined to upgrade.

The **average lifetime spend** represents the total amount a member has spent on purchases over their membership duration and is moderately correlated with the `ispremier` status. Members with Premier status tend to have a higher average lifetime spend compared to Insider members, indicating that those who upgrade to Premier generally engage more with the membership offerings, making more frequent or higher-value purchases. This trend suggests that spend behavior is a significant factor in predicting Premier membership, as increased spending likely reflects loyalty and higher engagement, qualities incentivized in the Premier tier. The correlation between `ispremier` and `life_spend_amt` (0.2203) underscores the relationship between cumulative spending and the likelihood of achieving or upgrading to Premier status.

2. **Are older**: Premier members are slightly older on average, although age alone isn't a strong predictor. Younger customers (especially those in the 13 to 21 age range) are less likely to upgrade to Premier.

3. **Engage more**: Customers who use rewards and promotions, like mobile and SMS offers, are more likely to convert to Premier.

4. **Have higher income and home values**: Customers with higher income and better home values are more likely to be Premier members.

5. **Purchase more tickets and concessions**: The more tickets and concessions a customer buys, the more likely they are to become Premier members.

In summary, customers who spend more, engage actively with offers, and have a higher income are more likely to convert to Premier membership.

Premier customers are typically high-value members who spend more and engage more with the services. They tend to be middle-aged adults, around 49 years old on average, and have a wide age range from as young as 8 to as old as 96. However, most of them are between 38 and 58 years old. Overall, Premier customers are loyal and have higher spending habits, valuing the extra perks and exclusive offers that come with being a Premier member.

#### Who are the customers that utilize Insider?

The customers who utilize the "Insider" membership are generally lower spenders compared to those in the "Premier" category. These customers tend to spend less on tickets, concessions, and rewards, indicating that they engage less frequently or make smaller purchases. Most Insider members have a relatively modest lifetime spend and accumulate fewer rewards over time. These members are usually younger and may not prioritize premium services, which are the main appeal of the "Premier" tier. In short, Insider customers are those who make less frequent or smaller purchases compared to the more engaged and higher-spending Premier customers.
